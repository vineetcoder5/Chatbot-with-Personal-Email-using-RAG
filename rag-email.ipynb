```python
file_path = "/kaggle/input/email-dataset/Untitled spreadsheet - mail (1).csv"
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
def combine_sentense(row):
    # Split the sentence into words
    return "date and time: "+str(row['data'])+"\nsubject: "+str(row['subject'])+"\nfrom: "+str(row['email'])+"\n\n"+row['body'] 

df['complete_email'] = df.apply(combine_sentense, axis=1)
    
```

```python
df['complete_email'][10]
```

```python
import google.generativeai as genai
```

```python
genai.configure(api_key="YOUR_API_KEY")
```

```python
def embed_text(text):
    result = genai.embed_content(model="models/text-embedding-004", content=text['complete_email'][:2000])
    return result['embedding']
    
```

```python
df['embedding'] = df.apply(embed_text, axis=1)
```

```python
df.to_pickle("email_embed.pkl")
```

```python
def distance(text):
    return np.dot(text['embedding'],text['question'])
```

```python
genai.configure(api_key="YOUR_API_KEY")

# Create the model
# See https://ai.google.dev/api/python/google/generativeai/GenerativeModel
generation_config = {
  "temperature": 1,
  "top_p": 0.95,
  "top_k": 64,
  "max_output_tokens": 100000,
  "response_mime_type": "text/plain",
}

model = genai.GenerativeModel(
  model_name="gemini-1.5-flash",
  generation_config=generation_config,
  # safety_settings = Adjust safety settings
  # See https://ai.google.dev/gemini-api/docs/safety-settings
  system_instruction= 'You are a personal assistant who is helping vineet to resond to question based on the emails',
)
```

```python
chat_session = model.start_chat(
  history=[
  ]
)

```

```python
def query_answer(question,df,k=5):
    question_embd = genai.embed_content(model="models/text-embedding-004", content=question)
    emb = question_embd['embedding']
    def questio(row):
        return emb
    df['question'] = df.apply(questio, axis=1)
    df['distance'] = df.apply(distance, axis=1)
    sorted_df = df.sort_values('distance', ascending=False, inplace=False)
    def combine_email(k=5):
        text = ""
        for i in range(k):
            text += sorted_df['complete_email'].iloc[i]+'\n\n'
        return text
    question_model = question+"\n\n"+"use this informattion from vineet email as context to answer the question\n"+combine_email(k)+"\n .please stick to this context to answer the question and only give answer dont give information where i can find it" 
    response = chat_session.send_message(question_model)
    return response.text
print(query_answer("who is dbms teacher",df,10))
```
